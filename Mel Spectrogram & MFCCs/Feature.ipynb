```python
import librosa
import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt
```

# Loading Audio using librosa

```python
file_path = 'vulture.mp3'
y, sr = librosa.load(file_path,sr=None) 
```

```python
y,sr
```

# Mel Filter Banks

```python
filter_banks = librosa.filters.mel(n_fft = 2048, sr=sr, n_mels=10)
```

```python
filter_banks.shape
```

```python
plt.figure(figsize=(15, 10))
librosa.display.specshow(filter_banks, 
                         sr=sr, 
                         x_axis="linear")
plt.colorbar(format="%+2.f")
plt.show()
```

# Extracting Mel Spectrogram

```python
mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=512, n_mels=10)
```

```python
mel_spectrogram,mel_spectrogram.shape
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(mel_spectrogram, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
log_mel_spectrogram.shape
```

# Log Mel_Spectrogram

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(log_mel_spectrogram, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

# Extracting MFCCs

```python
mfccs = librosa.feature.mfcc(y=y,sr=sr)
```

```python
mfccs,mfccs.shape
```

# Visualizing MFCCs

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(mfccs, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```
